# Kernel SVM

```python
import math

import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVC

from machine_learning.datasets.visualisation import plot_decision_regions
```

```python
np.random.seed(1)
```

## Apply an RBF-kernel SVM to a XOR dataset

```python
X_xor = np.random.randn(200, 2)
y_xor = np.where(
    np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0),
    1,
    -1
)
```

```python
plt.scatter(
    X_xor[y_xor == 1, 0],
    X_xor[y_xor == 1, 1],
    c = 'b',
    marker = 'x',
    label = '1'
)
plt.scatter(
    X_xor[y_xor == -1, 0],
    X_xor[y_xor == -1, 1],
    c = 'r',
    marker = 's',
    label = '-1'
)
plt.legend(loc = 'best')
plt.show()
```

```python
svm = SVC(kernel = 'rbf', random_state = 1, gamma = 0.1, C = 10)
svm.fit(X_xor, y_xor)
```

```python
plot_decision_regions(X_xor, y_xor, svm)
plt.legend(loc = 'upper left')
plt.show()
```

## Radial basis function

```python
gamma = 0.5
K = lambda distance: math.exp( - gamma * math.fabs(distance))
K_vectorised = np.vectorize(K)
```

```python
x1 = np.random.randn(200, 1)
x2 = np.random.randn(200, 1)
distance = x2 - x1
k = K_vectorised(distance)
```

```python
plt.scatter(distance, k)
```
